```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
movie_data=pd.read_csv("movies_metadata.csv")
movie_data.head()
```

```python
movie_data=movie_data.drop('homepage',axis=1)
movie_data=movie_data.drop('id',axis=1)
movie_data=movie_data.drop('poster_path',axis=1)
movie_data=movie_data.drop('belongs_to_collection',axis=1)
movie_data=movie_data.drop('overview',axis=1)
movie_data=movie_data.drop('production_countries',axis=1)
movie_data=movie_data.drop('production_companies',axis=1)
movie_data['revenue']=movie_data['revenue']/10000000
movie_data=movie_data.loc[movie_data['status']=='Released']
movie_data=movie_data.drop('budget',axis=1)
movie_data=movie_data.drop('spoken_languages',axis=1)
movie_data=movie_data.drop('tagline',axis=1)
movie_data=movie_data.drop('genres',axis=1)
movie_data=movie_data.drop('release_date',axis=1)
movie_data=movie_data.drop('video',axis=1)
movie_data=movie_data.drop('status',axis=1)
movie_data=movie_data.drop('original_title',axis=1)
movie_data['popularity']=pd.to_numeric(movie_data['popularity'])
movie_data=movie_data.loc[movie_data['runtime']<250.0]
movie_data=movie_data.loc[movie_data['popularity']>0.5]

movie_data['vote_average_count']=movie_data['vote_average']*movie_data['vote_count']
movie_data['vote_average_count']=movie_data['vote_average_count']/(max(movie_data['vote_average_count'])-min(movie_data['vote_average_count']))

movie_data['popularity']=movie_data['popularity']/(max(movie_data['popularity'])-min(movie_data['popularity']))
movie_data['revenue']=movie_data['revenue']/(max(movie_data['revenue'])-min(movie_data['revenue']))
movie_data['vote_average']=movie_data['vote_average']/(max(movie_data['vote_average'])-min(movie_data['vote_average']))
movie_data['vote_count']=movie_data['vote_count']/(max(movie_data['vote_count'])-min(movie_data['vote_count']))
movie_data['final_rating']=100*(50*movie_data['popularity']+20*movie_data['revenue']+30*movie_data['vote_average_count'])

movie_children_data=movie_data.loc[movie_data['adult']=='False']
movie_children_data=movie_children_data.drop('adult',axis=1)

movie_eng_child=movie_children_data.loc[movie_children_data['original_language']=='en']
movie_eng_child=movie_children_data.drop('original_language',axis=1)

movie_noneng_child=movie_children_data.loc[movie_children_data['original_language']!='en']
movie_noneng_child=movie_children_data.drop('original_language',axis=1)



movie_data.head()
```
